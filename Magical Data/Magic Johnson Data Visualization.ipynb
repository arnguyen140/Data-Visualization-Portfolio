```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Magic_Johnson_Totals_Stats.csv')
df
```

```python
season = []
fieldAtmpt = []
assists = []
for obj in df['Season']:
    season.append(obj)
for obj in df['FGA']:
    fieldAtmpt.append(obj)
for obj in df['AST']:
    assists.append(obj)
```

```python
plt.plot(season,fieldAtmpt, label = 'Field Goal Attemps')
plt.plot(season,assists, label = 'Assists')
plt.title('Magic Johnson Career Field Goal Attempts vs Assists')
plt.legend()
plt.xticks(rotation=90)
plt.xlabel('Season')
plt.show()
```
