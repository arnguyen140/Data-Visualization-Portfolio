```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(8)
y1 = [13.7,6.7,32.3,29,11,3,2,1.5]
y2 = [20.1,8.9,27.6,26,10,4,3,2]
y3 = [20.6,7,16.2,29,12,5,3,4]
width = 0.2
plt.rcParams['figure.figsize'] = [20, 10]
plt.bar(x - 0.2, y1, width, align='center', color='#8ecae6')
plt.bar(x, y2, width, align='center', color='#219ebc')
plt.bar(x + 0.2, y3, width, align='center', color='#023047')
plt.xticks(x, ['Age 0-9', 'Age 10-19', 'Age 20-29', 'Age 30-39', 'Age 40-49', 'Age 50-59','Age 60-69', 'Age 70 and over'])
plt.title('Age Distribution of Influenza and Pneumonia Deaths in Boston 1918')
plt.ylabel('Percent of Deaths')
plt.legend(['September', 'October', 'November'])
plt.show()
```
